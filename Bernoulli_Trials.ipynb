```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_bernoulli_trials(N, p, num_simulations):
    """
    Simulate Bernoulli trials and study their values.

    Parameters:
    - N: int, number of trials
    - p: float, probability of success
    - num_simulations: int, number of simulations to run

    Returns:
    - results: numpy array, number of successes in each simulation
    """
    # Simulate Bernoulli trials with 'H' for heads (success) and 'T' for tails (failure)
    trials = np.random.choice(['H', 'T'], size=(num_simulations, N), p=[p, 1-p])
    
    # Convert 'H' to 1 and 'T' to 0, then sum the number of successes in each simulation
    trials_numeric = np.where(trials == 'H', 1, 0)
    results = trials_numeric.sum(axis=1)
    
    return results

def plot_results(results, N):
    """
    Plot the results of the Bernoulli trials simulation.

    Parameters:
    - results: numpy array, number of successes in each simulation
    - N: int, number of trials
    """
    plt.hist(results, bins=np.arange(N + 2) - 0.5, edgecolor='black', density=True)
    plt.xlabel('Number of successes')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of Bernoulli Trials Results (N={N})')
    plt.xticks(np.arange(N + 1))
    plt.show()

# Parameters
p = 0.5  # probability of success (heads)
num_simulations = 10000  # number of simulations

# Simulate and plot for different values of N
for N in [2, 3, 4]:
    results = simulate_bernoulli_trials(N, p, num_simulations)
    print(f"Results for N={N}:")
    print(f"Mean number of Heads(Hs): {np.mean(results)}")
    # print(f"Standard deviation of successes: {np.std(results)}")
    plot_results(results, N)
```
