```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
```

**Experiment**: Tossing 10 coins simultaneously. <br>
**Observation**: X is a variable that keeps track of the number H's in one single experiment. <br>

Now, X is a random variable. Why? <br>

Because everytime you do the experiment, X can take different values in the range 0–10. Let's try to run the python program below based on it.

### Repeating an experiment of 10 coin tosses

```python
# Number of coin tosses in each experiment
num_tosses = 10

# Number of experiments
num_repititions = 10000

# the choices in a coin toss
choices = ['H','T']

#an array to store value of X in each repitition
X = np.zeros((num_repititions))

# Simulate the experiments
print("Values of X")
for i in range(num_repititions):
    toss_results = np.random.choice(choices, size=num_tosses)
    X[i] = int(list(toss_results).count('H'))
    print ( "Repitition %3i is %3i"%(i+1, X[i]) )
    
print("So all values of X --->", X.astype('int'))
```

```python
def theoretical_distribution(X, Ntosses, p=0.5):
    """
    Calculate the theoretical binomial distribution.
    
    Parameters:
    X (int or array-like): Number of successes (can be a single integer or an array of integers)
    Ntrials (int): Number of trials
    p (float): Probability of success in each trial (default is 0.5)
    
    Returns:
    array-like: A list or array of probabilities for the given number of successes
    """
    X = np.asarray(X)
    binom_coeff = factorial(Ntosses) / (factorial(X) * factorial(Ntosses - X))
    probability = binom_coeff * (p ** X) * ((1 - p) ** (Ntosses - X))
    return probability
```
